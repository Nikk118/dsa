```python
class Node:
    def __init__(self,value):
        self.value=value
        self.next=None
        
```

```python
class linkedlist:
    def __init__(self):
        self.head=None
        self.tail=None
        self.n=0

    def add(self,value):
        
        n=Node(value)
        n.next=self.head
        self.head=n
        if self.n==0:
            self.tail=n
            self.head=n
          

        self.n=self.n+1
        
    # insert in middle 
    def insert(self,ref,value):
        n=Node(value)
        curr=self.head
        flag=True
        while(curr!=self.tail):
            if curr.value==ref:
                n.next=curr.next
                curr.next=n
                flag=False
                self.n=self.n+1
                return "successfully inserted"
            curr=curr.next
        if flag:
            return "this number is not in the linked list"



    def append(self,value):
        n=Node(value)
        if self.n==0:
            self.tail=n
            self.head=n
        self.tail.next=n
        self.tail=n
        self.n+=1

    def clear(self):
        self.head=None
        self.n=0

    def pop_head(self):
        self.head=self.head.next
        self.n-=1

    def pop(self):

        if self.head is None:
            return "list is empty"
        
        if self.n==1:
            self.head=None
            self.n-=1
            print(self.n)
            return
        
        curr=self.head
        while curr.next.next!=None:
            curr=curr.next
        curr.next=None
        self.tail=curr
        self.n-=1
        print(self.n)

    def search(self,value):
        curr=self.head
        pos=0
        while curr.next!=None:
            if curr.value==value:
                return pos
            pos+=1
            curr=curr.next
        return "not found"



    def remove(self,ref):
        if self.head==None:
            return "empty list"
        
        curr=self.head

        if ref==self.head.value:
            self.pop_head()
        
        while(curr.next!=None):
            if curr.next.value==ref:
               break
            curr=curr.next
            
        if curr.next==None:
            return "not found"
        else:
            curr.next=curr.next.next
            self.n-=1

    def __str__(self):
        curr=self.head
        result=""
        while(curr!=None):
            result+=str(curr.value)+"->"
            curr=curr.next
        return "["+result[:-2]+"]"

    def __len__(self):
        return self.n
    

    def __delitem__(self, key):
        if 0<= key < self.n:
            curr=self.head
            if key==0:
                self.pop_head()
                return
            for i in range(key):
                if i==key-1:
                    curr.next=curr.next.next   
                    curr=curr.next
        raise IndexError("index out of range")
           
        
            




        
```
