```python
import ctypes
```

```python
class meralist:
    def __init__(self):
        self.n=0
        self.size=1

        self.A=self.__make_array(self.size)

    def __len__(self):
        return self.n

    def append(self,item):
        if self.size==self.n:
            self.__resize(self.size*2)

        self.A[self.n]=item
        self.n=self.n+1


    def __str__(self):
        result=""
        if self.n==0:
            return "nothing here"
        for i in range(self.n):
            result=result+str(self.A[i])+","

        return "["+result[:-1]+"]"


    def __getitem__(self, key):
        if key>=self.n or key<0:
            return "out of bound"
        return self.A[key]

    def pop(self):
        if self.n==0:
            return "empty list"
        print(self.A[self.n-1])
        self.n=self.n-1

    def clear(self):
        if self.n==0:
            return "cleared"
        self.n=0
        return "cleared"

    def find(self,item):
        for i in range(self.n):
            if self.A[i]==item:
                return i
            
        return "not found"

    def insert(self,index,item):
        if self.size==self.n:
            self.__resize(self.n*2)
        
        for i in range(self.n,index-1,-1):
            self.A[i]=self.A[i-1]

        self.A[index]=item
        self.n=self.n+1


    def remove(self,item):
        index=self.find(item)
        self.__delitem__(index)

    def __resize(self,new_capacity):
        B=self.__make_array(new_capacity)
        self.size=new_capacity

        for i in range(self.n):
            B[i]=self.A[i]

        self.A=B 
        
    def __delitem__(self, key):
        if key<self.n:
            for i in range(key,self.n-1):
                self.A[i]=self.A[i+1]
            self.n=self.n-1
    def sum(self):
        sum=0
        for i in range(self.n):
            sum+=self.A[i]
        return sum

    def __make_array(self,capacity):
        return (capacity*ctypes.py_object)()
```
