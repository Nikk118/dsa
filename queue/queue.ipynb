```python
class Node:
    def __init__(self,value):
        self.value=value
        self.next=None
```

```python
class Queue:
    def __init__(self):
        self.head=None
        self.tail=None
        self.n=0

    def isEmpty(self):
        return self.n==0

    def enqueue(self,value):
        n=Node(value)
        if self.head==None:
            self.head=n
            self.tail=n
        else:
            self.tail.next=n
            self.tail=n
        self.n+=1

    def dequeue(self):
        if self.n==0:
            return "empty queue"
        self.head=self.head.next
        self.n-=1
      
    def __str__(self):
        res=""
        curr=self.head
        while curr!=None:
            res+=str(curr.value) +" "
            curr=curr.next
        return res
```

```python
q=Queue()
```

```python
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
```

```python
q.isEmpty()
```

```python
print(q)
```

```python
q.dequeue()
```

```python
class CircularQueue:
    def __init__(self,size):
        self.capacity=size
        self.rear=-1
        self.front=0
        self.q=np.empty(self.capacity)
        self.items=0
    def enqueue(self,value):
        if self.items==self.capacity:
            return "full"
        else:
            self.rear=(self.rear+1)%self.capacity
            self.q[self.rear]=value
            self.items+=1

    def traverce(self):
        if self.items==0:
            return "empty queue"
        temp=self.front
        for i in range(self.rear+1):
            if temp!=self.rear+1:
                print(self.q[temp] ,end=" ")
                temp=temp+1%self.capacity

    def dequeue(self):
        if self.items==0:
            return "empty queue"
        print(self.q[self.front])
        self.front=(self.front+1)%self.capacity
        self.items-=1


            
        
```
